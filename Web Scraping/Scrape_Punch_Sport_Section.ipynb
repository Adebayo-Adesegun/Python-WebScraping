```python
# importing the necessary packages
import requests
from bs4 import BeautifulSoup

url = 'https://punchng.com/topics/sports/'

r1 = requests.get(url)
coverpage = r1.content

print(coverpage)
```

```python

soup1 = BeautifulSoup(coverpage, 'html5lib')
coverpage_news = soup1.find_all('div', class_='items col-sm-12')
print(coverpage_news)
```

```python
#Check the number of articles seen
print(len(coverpage_news))
```

```python
#Loop Through Coverage_news 
list_of_links = []
articles_contents = []
articles_titles = []


coverage_news_length = len(coverpage_news)

for i in range(0,coverage_news_length):
    title = coverpage_news[i].find('h2').get_text()
    link = coverpage_news[i].find('a')['href']
    list_of_links.append(link)
    articles_titles.append(title)
    
    print(list_of_links)
    print(articles_titles)
```

```python
#making requests for the article contents 
# Reading the content (it is divided in paragraphs)

news_contents = []
for article_link in list_of_links:
    article = requests.get(article_link)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')
    body = soup_article.find_all('div', class_='entry-content')
    x = body[0].find_all('p')
    
    #unifying the Paragraphs
    list_paragraphs = []
    for p in range(0, len(x)):
        paragraph = x[p].get_text()
        list_paragraphs.append(paragraph)
        final_article = " ".join(list_paragraphs)
        news_contents.append(final_article)
    

print(news_contents)
    
    
    
    

   
```
